# Example code for binning continuous data

#### A continuous target variable can be binned into discrete class labels, for example using `pd.cut()` or `pd.qcut()`

```python
import numpy as np
import pandas as pd

# Generate a numeric continous normal-distributed target variable
df = pd.DataFrame(np.random.normal(0, 1, 10000), columns=['target'])

# Plot the target variable distribution
df['target'].plot.hist(bins=50).set_title('Target variable distribution')
```

## Example using equally spaced bins with pd.cut

```python
# Using bins with equal value range for each bin but possibly different amount of samples per bin (here: 10 bins)
########################################################################################
num_bins = 10
df['binned_target'] = pd.cut(df['target'], num_bins, labels=False)
########################################################################################

# Check that it worked (and show a bit what it does):

# Print unique values of the binned target variable (these are the class labels)
print(sorted(df['binned_target'].unique()))

# Plot class histogram
df['binned_target'].plot.hist(rwidth=0.9) # rwidth is the width of the bars in the histogram plot
```

## Example using equally populated bins with pd.qcut

```python
# Using bins with equal amount of samples per bin but variable value range for each bin (here: 10 bins)
########################################################################################
num_bins = 10
df['binned_target'] = pd.qcut(df['target'], num_bins, labels=False)
########################################################################################

# Check that it worked (and show a bit what it does):

# Print unique values of the binned target variable (these are the class labels)
print(sorted(df['binned_target'].unique()))

# Plot class histogram
df['binned_target'].plot.hist(rwidth=0.9) # rwidth is the width of the bars in the histogram plot
```
